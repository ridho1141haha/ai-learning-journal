```python
nilai = [80, 75, 90, 60, 85]

def evaluasi_nilai(data):
    hasil = []
    for n in data:
        if n >= 75:
            hasil.append("LULUS")
        else:
            hasil.append("TIDAK LULUS")
    return hasil

print(evaluasi_nilai(nilai))
```
