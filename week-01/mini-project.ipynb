```python
import pandas as pd
import numpy as np

data = {
    "nama": ["Andi", "Budi", "Citra", "Dewi", "Eka", "Fajar", "Budi"],
    "nilai": [80, None, 90, "70", 85, "tidak ikut", None],
    "kelas": ["A", "A", "B", None, "B", "A", "A"]
}

df = pd.DataFrame(data)
df
## disini kita cuma di suruh untuk merapikan data saja
# jadi kita akan menghapus data yang tidak valid pada kolom nilai
```

```python
df.info()
df.isnull().sum()
df.head()
```

```python
df["nilai"] = pd.to_numeric(df["nilai"], errors="coerce")
```

```python
df["nilai"] = df["nilai"].fillna(df["nilai"].mean())
df["kelas"] = df["kelas"].fillna("Unknown")
```

```python
df = df.drop_duplicates()
```

```python
df.loc[:, "status"] = df["nilai"].apply(
    lambda x: "LULUS" if x >= 75 else "TIDAK LULUS"
)
```

```python
df["nilai"].mean()
```

```python
df.groupby("kelas")["nilai"].mean()
```
